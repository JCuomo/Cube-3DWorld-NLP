```python
import multiprocessing
import time
import class_plot3Dsolution
import sys
test_n = sys.argv[1]
```

```python
def part1(test_n): 
    %run -i 'part1_findGoal.py' $test_n
def part2(test_n):
    %run -i 'part2_findStackMoves.py' $test_n
def part3(test_n): 
    %run -i 'part3_findPaths.py' $test_n
```

```python
def timeLoops(foo, testNum, runningTime=10):
    stop = False
    while not stop:
        # Start foo as a process
        p = multiprocessing.Process(target=foo, name="part1", args=(testNum,))
        p.start()
        stop = True
        # Wait a maximum of 10 seconds for foo
        p.join(runningTime) # Usage: join([timeout in seconds])
        # If thread is active
        if p.is_alive():
            stop = False
            runningTime += runningTime
            print ("Restarting and running for",runningTime, "sec")
            p.terminate()  # Terminate foo
            p.join()
```

```python
timeLoops(part1, test_n)
```

```python
part2(test_n)
```

```python
timeLoops(part3, test_n)
```
